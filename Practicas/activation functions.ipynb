#  Dibujo de la función f(x)

Se graficarán algunas de las funciones de activación más importantes.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plot method
def plot_function(x, y, title):
    plt.figure(figsize=(5,4))
    plt.title(title)
    plt.plot(x, y, linewidth=3, color='darkblue')
    # plot 0 line
    plt.plot(x, np.zeros(100), linewidth=1, color='black')

    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.show()
```

## Hard limit
Su ecuación es:
$$f(x) = \left\{
\begin{array}{ll}
0 & \text{si } x < 0 \\
1 & \text{si } x \geq 0
\end{array}
\right.$$

```python
# Graficamos la función hardlim
x = np.linspace(-2, 2, 100)
y = np.heaviside(x, 1)

plot_function(x, y, 'Hard Limit')
```

# Symmetrical hard limit
Su ecuación es:
$$
f(x) = \left\{
\begin{array}{ll}
-1 & \text{si } x < 0 \\
1 & \text{si } x \geq 0
\end{array}
\right.
$$

```python
# Graficamos la función symmetrical hardlim
y_1 = np.heaviside(x, 1)
y_2 = np.heaviside(x, 1) - 1
y = y_1 + y_2

plot_function(x, y, 'Symmetrical Hard Limit')
```

## Linear
Su ecuación es:
$$
f(x) = x
$$

```python
y = x

plot_function(x, y, 'Linear')
```

## Saturation linear
Su ecuación es:
$$
f(x) = \left\{
\begin{array}{ll}
0 & \text{si } x < 0 \\
x & \text{si } 0 \leq x \leq 1 \\
1 & \text{si } x > 1
\end{array}
\right.
$$

```python
# Graficamos la función saturación no simétrica
x = np.linspace(-2, 3, 100)
y = np.minimum(1, np.maximum(0, x))

plot_function(x, y, 'Saturation')
```

## Symmetrical saturation linear
Su ecuación es:
$$
f(x) = \left\{
\begin{array}{ll}
-1 & \text{si } x < -1 \\
x & \text{si } -1 \leq x \leq 1 \\
1 & \text{si } x > 1
\end{array}
\right.
$$

```python
x = np.linspace(-2, 2, 100)
y = np.clip(x, -1, 1)

plot_function(x, y, 'Symmetrical Saturation')
```

## Log-sigmoid
Su ecuación es:
$$
f(x) = \frac{1}{1 + e^{-x}}
$$

```python
x = np.linspace(-5, 5, 100)
y = 1 / (1 + np.exp(-x))

plot_function(x, y, 'Log-sigmoid')
```

## Hyperbolic tangent sigmoid
Su ecuación es:
$$
f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$


```python
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

plot_function(x, y, 'Hyperbolic Tangent Sigmoid')
```

## Positive linear
Su ecuación es:
$$
f(x) = \left\{
\begin{array}{ll}
0 & \text{si } x < 0 \\
x & \text{si } x \geq 0
\end{array}
\right.
$$

```python
# Positive linear
x = np.linspace(-2, 2, 100)
y = np.maximum(0, x)

plot_function(x, y, 'Positive Linear')
```

## Competitive
Su ecuación es:
$$
f(x) = \left\{
\begin{array}{ll}
1 & \text{si } x = \max(x_1, x_2, \ldots, x_n) \\
0 & \text{en otro caso}
\end{array}
\right.
$$


```python
# Competitive
# plot frecuency like max function
x = [0, 1, 2, 3, 4, 5]
y = [0, 0, 0, 0, 0, 1]

plt.figure(figsize=(5,4))
plt.title('Competitive')
plt.plot(x, y, linewidth=3, color='darkblue', marker='o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

## LReLU
Su ecuación es:
$$
f(x) = \left\{
\begin{array}{ll}
\alpha x & \text{si } x < 0 \\
x & \text{si } x \geq 0
\end{array}
\right.
$$



```python
# LReLU
x = np.linspace(-2, 2, 100)
y = np.maximum(0.1 * x, x)

plot_function(x, y, 'Leaky Rectified Linear Unit')
```

## Swish
Su ecuación es:
$$
f(x) = \frac{x}{1 + e^{-x}}
$$


```python
# swish
x = np.linspace(-5, 5, 100)
y = x / (1 + np.exp(-x))

plot_function(x, y, 'Swish')
```

## Softplus
Su ecuación es:
$$
f(x) = \ln(1 + e^x)
$$

```python
#  Softplus
x = np.linspace(-5, 5, 100)
y = np.log(1 + np.exp(x))

plot_function(x, y, 'Softplus')
```

## ELU
Su ecuación es:
$$
f(x) = \left\{
\begin{array}{ll}
\alpha (e^x - 1) & \text{si } x < 0 \\
x & \text{si } x \geq 0
\end{array}
\right.
$$


```python
# elu
x = np.linspace(-5, 5, 100)
y = np.maximum(0, x) + np.minimum(0, np.exp(x) - 1)

plot_function(x, y, 'Exponential Linear Unit')
```

## SELU
Su ecuación es:
$$
f(x) = \left\{
\begin{array}{ll}
\lambda \alpha (e^x - 1) & \text{si } x < 0 \\
\lambda x & \text{si } x \geq 0
\end{array}
\right.
$$


```python
# selu
x = np.linspace(-5, 5, 100)
y = 1.0507 * (np.maximum(0, x) + np.minimum(0, np.exp(x) - 1))

plot_function(x, y, 'Scaled Exponential Linear Unit')
```

## GELU
Su ecuación es:
$$
\begin{eqnarray}
f(x) &=& x P(X \leq x)=x \Phi(x) \\ &=& \frac{x}{2} \left(1 + erf\left(\frac{x}{\sqrt{2}}\right)\right)
\end{eqnarray}
$$

```python
## GELU
x = np.linspace(-5, 5, 100)
y = x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3))))

plot_function(x, y, 'Gaussian Error Linear Unit')
```

## Mish
Su ecuación es:
$$
f(x) = x \tanh(\ln(1 + e^x))
$$

```python
x = np.linspace(-5, 5, 100)
y = x*np.tanh(np.log(1+np.exp(x)))

plot_function(x, y, 'Mish')
```

## ReLU6
Su función:
$$
f(x) = \left\{
\begin{array}{ll}
0 & \text{si } x < 0 \\
x & \text{si } 0 \leq x \leq 6 \\
6 & \text{si } x > 6
\end{array}
\right.
$$


```python
# ReLU6
x = np.linspace(-8, 8, 100)
y = np.minimum(6, np.maximum(0, x))

plot_function(x, y, 'ReLU6')
```

## Hard swish
Su ecuación es:
$$
f(x) = x\frac{ReLU6(x+3)}{6}
$$


```python
def RELU6(x):
    return np.minimum(6, np.maximum(0, x))
```

```python
# hard swish
x = np.linspace(-5, 5, 100)
y = x * RELU6(x + 3) / 6

plot_function(x, y, 'Hard Swish')
```

## SiLU
Su ecuación es:
$$
f(x) = x \cdot \sigma(x)
$$

```python
## SiLU
x = np.linspace(-5, 5, 100)
y = x / (1 + np.exp(-x))

plot_function(x, y, 'SiLU')
```

## LiSA
Su ecuación es:
$$
f(x) = \left\{
\begin{array}{ll}
\alpha_1 x - \alpha_1 + 1, & \text{si } x > 1 \\
x, & \text{si } 0 \le x \le 1 \\
\alpha_2 x , & \text{si } x < 0
\end{array}
\right.
$$


```python
## LiSA
alpha_1 = 0.2
alpha_2 = 0.2

x_1 = np.linspace(-5, 0, 35)
x_2 = np.linspace(0, 1, 30)
x_3 = np.linspace(1, 5, 35)

y_3 = alpha_1 * x_3 - alpha_1 + 1
y_2 = x_2
y_1 = alpha_2 * x_1

x = np.concatenate((x_1, x_2, x_3))
y = np.concatenate((y_1, y_2, y_3))

plot_function(x, y, 'LiSA')
```
